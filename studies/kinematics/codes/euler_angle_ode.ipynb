```python
%%octave
1;

function dydt = func(t, y)
    C = [0, sin(y(3)), cos(y(3));
        0, cos(y(3))*cos(y(2)), -sin(y(3))*cos(y(2));
        cos(y(2)), sin(y(3))*sin(y(2)), cos(y(3))*sin(y(2))] / cos(y(2));
    omega = 20 * [sin(0.1*t);
            0.01;
            cos(0.1*t)] * (pi/180);
    dydt = C * omega;

end

% Euler angles at 60s
tspan = [0, 42, 60];
y0 = [40; 30; 80] * (pi/180);
[t, y] = ode45(@func, tspan, y0);
disp([t(end), y(end, 1), y(end, 2), y(end, 3)]);

% Euler angles at 42s
psi = y(2, 1);
theta = y(2, 2);
phi = y(2, 3);
norm = sqrt(psi^2 + theta^2 + phi^2);
disp(norm);
```
