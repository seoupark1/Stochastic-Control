```python
%%octave

function dbdt = sys(t, b)

    %% angular velocity
    omega1 = sin(0.1*t); omega2 = 0.01; omega3 = cos(0.1*t);
    omega = 20 * [omega1; omega2; omega3] * (pi/180);

    DCM = [-b(2), -b(3), -b(4);
           b(1), -b(4), b(3);
           b(4), b(1), -b(2);
           -b(3), b(2), b(1)];

    dbdt = 1/2 * DCM * omega;

end


%% time
tspan = [0, 42];

%% initial quaternion
b0 = [0.408248; 0.; 0.408248; 0.816497];

[t, b] = ode45(@sys, tspan, b0);
b0= b(end, 1); b1 = b(end, 2); b2 = b(end, 3); b3 = b(end, 4);
disp([t(end), b0, b1, b2, b3]);

%% norm
norm = sqrt(b1^2 + b2^2 + b3^2);
disp(norm);
```
