```python
import math

def result():
    theta1 = math.radians(10)
    theta2 = math.radians(20)
    theta3 = math.radians(30)

    c_31 = math.cos(theta3)*math.sin(theta2)*math.cos(theta1) + math.sin(theta3)*math.sin(theta1)
    c_32 = math.cos(theta3)*math.sin(theta2)*math.sin(theta1) - math.sin(theta3)*math.cos(theta1)
    c_33 = math.cos(theta3)*math.cos(theta2)
    c_13 = -math.sin(theta2)
    c_23 = math.sin(theta3)*math.cos(theta2)

    theta1 = math.atan2(c_31, -c_32)
    theta2 = math.acos(c_33)
    theta3 = math.atan2(c_13, c_23)

    eulerAngles = [math.degrees(theta1), math.degrees(theta2), math.degrees(theta3)] # degrees
    return eulerAngles

print(result())
```

```python
import math
import numpy as np

def result():

    theta1 = math.radians(10)  # BN
    theta2 = math.radians(20)
    theta3 = math.radians(30)
    BN1 = np.array([math.cos(theta2)*math.cos(theta1), math.cos(theta2)*math.sin(theta1), -math.sin(theta2)])
    BN2 = np.array([math.sin(theta3)*math.sin(theta2)*math.cos(theta1) - math.cos(theta3)*math.sin(theta1), math.sin(theta3)*math.sin(theta2)*math.sin(theta1) + math.cos(theta3)*math.cos(theta1), math.sin(theta3)*math.cos(theta2)])
    BN3 = np.array([math.cos(theta3)*math.sin(theta2)*math.cos(theta1) + math.sin(theta3)*math.sin(theta1), math.cos(theta3)*math.sin(theta2)*math.sin(theta1) - math.sin(theta3)*math.cos(theta1), math.cos(theta3)*math.cos(theta2)])
    BN = np.array([BN1, BN2, BN3])

    theta1 = math.radians(-5)  # RN
    theta2 = math.radians(5)
    theta3 = math.radians(5)
    RN1 = np.array([math.cos(theta2)*math.cos(theta1), math.cos(theta2)*math.sin(theta1), -math.sin(theta2)])
    RN2 = np.array([math.sin(theta3)*math.sin(theta2)*math.cos(theta1) - math.cos(theta3)*math.sin(theta1), math.sin(theta3)*math.sin(theta2)*math.sin(theta1) + math.cos(theta3)*math.cos(theta1), math.sin(theta3)*math.cos(theta2)])
    RN3 = np.array([math.cos(theta3)*math.sin(theta2)*math.cos(theta1) + math.sin(theta3)*math.sin(theta1), math.cos(theta3)*math.sin(theta2)*math.sin(theta1) - math.sin(theta3)*math.cos(theta1), math.cos(theta3)*math.cos(theta2)])
    RN = np.array([RN1, RN2, RN3])

    BR = BN @ RN.transpose()
    c_12 = BR[0, 1]  # BR(3-2-1 DCM)에서 inverse mappig을 위한 요소 뽑아내기
    c_11 = BR[0, 0]
    c_13 = BR[0, 2]
    c_23 = BR[1, 2]
    c_33 = BR[2, 2]

    phi1 = math.atan2(c_12, c_11)
    phi2 = -math.asin(c_13)
    phi3 = math.atan2(c_23, c_33)

    eulerAngles = [math.degrees(phi1), math.degrees(phi2), math.degrees(phi3) ] # degrees
    return eulerAngles

print(result())
```
