```python
import numpy as np

def result():
    # about beta_FB
    b0, b1, b2, b3 = 0.359211, 0.898027, 0.179605, 0.179605
    row1 = np.array([b0, -b1, -b2, -b3])
    row2 = np.array([b1, b0, b3, -b2])
    row3 = np.array([b2, -b3, b0, b1])
    row4 = np.array([b3, b2, -b1, b0])
    beta_FB = np.array([row1, row2, row3, row4])

    # about beta_BN
    b0, b1, b2, b3 = 0.774597, 0.258199, 0.516398, 0.258199
    beta_BN = np.array([b0, b1, b2, b3]).reshape(4,1)

    # shortest path
    beta_FN = beta_FB @ beta_BN
    if beta_FN[0] < 0:
        beta_FN = -beta_FN

    return beta_FN

print(result())
```

```python
import numpy as np

def result():
    # about beta_FN
    b0, b1, b2, b3 = 0.359211, 0.898027, 0.179605, 0.179605
    beta_FN = np.array([b0, b1, b2, b3]).reshape(4,1)

    # about beta_BN
    b0, b1, b2, b3 = -0.377964, 0.755929, 0.377964, 0.377964
    row1 = np.array([b0, -b1, -b2, -b3])
    row2 = np.array([b1, b0, -b3, b2])
    row3 = np.array([b2, b3, b0, -b1])
    row4 = np.array([b3, -b2, b1, b0])
    beta_BN = np.array([row1, row2, row3, row4])

    # shortest path
    beta_FB = beta_BN.transpose() @ beta_FN
    if beta_FB[0] < 0:
        beta_FB = -beta_FB

    return beta_FB

print(result())
```
