```python
import numpy as np

def tilde(v):
    result = np.array([[0, -v[2,0], v[1,0]],
                       [v[2,0], 0, -v[0,0]],
                       [-v[1,0], v[0,0], 0]])

    return result

def CRP2C(q):
    result = ((1 - np.vdot(q, q)) * np.eye(3) + 2 * q @ np.transpose(q) - 2 * tilde(q)) / (1 + np.vdot(q, q))

    return result

def OLAE(v1_est, v2_est, v1_true, v2_true):
    # normalization
    v1_est = v1_est / np.linalg.norm(v1_est)
    v2_est = v2_est / np.linalg.norm(v2_est)
    v1_true = v1_true / np.linalg.norm(v1_true)
    v2_true = v2_true / np.linalg.norm(v2_true)

    # sensor's weight
    w1 = 1
    w2 = 1

    # initial value
    d1 = v1_est - v1_true
    d2 = v2_est - v2_true
    s1 = v1_est + v1_true
    s2 = v2_est + v2_true

    # derive qbar
    d = np.block([[d1],
                  [d2]])
    s = np.block([[tilde(s1)],
                  [tilde(s2)]])
    w = np.block([[w1 * np.eye(3), np.zeros((3, 3))],
                  [np.zeros((3, 3)), w2 * np.eye(3)]])
    qbar = np.linalg.inv(np.transpose(s) @ w @ s) @ np.transpose(s) @ w @ d
    dcm_BbarN = CRP2C(qbar)

    return dcm_BbarN

v1_est = np.array([0.8273, 0.5541, -0.0920]).reshape(3,1)
v2_est = np.array([-0.8285, 0.5522, -0.0955]).reshape(3,1)
v1_true = np.array([-0.1517, -0.9669, 0.2050]).reshape(3,1)
v2_true = np.array([-0.8393,0.4494, -0.3044]).reshape(3,1)
print(OLAE(v1_est, v2_est, v1_true, v2_true))
```
