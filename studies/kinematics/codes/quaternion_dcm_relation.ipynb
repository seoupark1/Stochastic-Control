![problem 2](https://github.com/seoupark1/Stochastic-Control/blob/main/assets/images/kinematics/quaternions/quaternion_dcm_relation_02.png?raw=true)

```python
import numpy as np
import math

# given dcm
row1 = np.array([-0.529403, -0.467056, 0.708231])
row2 = np.array([-0.474115, -0.529403, -0.703525])
row3 = np.array([0.703525, -0.708231, 0.0588291])
dcm = np.array([row1, row2, row3])

def largest_val():
    # Sheppard's method 1st step: 소수점 오차를 줄이기 위해 가장 큰 값 찾기
    b0_square = (np.trace(dcm) + 1) / 4
    b1_square = (2*dcm[0,0] - np.trace(dcm) + 1) / 4
    b2_square = (2*dcm[1,1] - np.trace(dcm) + 1) / 4
    b3_square = (2*dcm[2,2] - np.trace(dcm) + 1) / 4

    data = np.array([b0_square, b1_square, b2_square, b3_square])
    names = ['b0_square', 'b1_square', 'b2_square', 'b3_square']
    max_index = np.argmax(data)
    result = names[max_index]

    return result

print(largest_val())
```

```python
def result():
    largest_value = (2*dcm[2,2] - np.trace(dcm) + 1) / 4
    b3 = math.sqrt(largest_value)

    # Sheppard's method 2nd step: 나머지 EPs 찾기
    b0 = (dcm[0,1] - dcm[1,0]) / (4*b3)
    b1 = (dcm[2,0] + dcm[0,2]) / (4*b3)
    b2 = (dcm[1,2] + dcm[2,1]) / (4*b3)

    # finding shortest path
    if b0 < 0:
        b0, b1, b2, b3 = -b0, -b1, -b2, -b3

    ep = [b0, b1, b2, b3]
    return ep

print(result())
```

```python
import numpy as np
import math

theta1 = math.radians(20)
theta2 = math.radians(10)
theta3 = math.radians(-10)

# 3-2-1 EA to dcm
row1 = np.array([math.cos(theta2)*math.cos(theta1), math.cos(theta2)*math.sin(theta1), -math.sin(theta2)])
row2 = np.array([math.sin(theta3)*math.sin(theta2)*math.cos(theta1) - math.cos(theta3)*math.sin(theta1), math.sin(theta3)*math.sin(theta2)*math.sin(theta1) + math.cos(theta3)*math.cos(theta1), math.sin(theta3)*math.cos(theta2)])
row3 = np.array([math.cos(theta3)*math.sin(theta2)*math.cos(theta1) + math.sin(theta3)*math.sin(theta1), math.cos(theta3)*math.sin(theta2)*math.sin(theta1) - math.sin(theta3)*math.cos(theta1), math.cos(theta3)*math.cos(theta2)])
dcm = np.array([row1, row2, row3])

def largest_val():
    b0_square = (np.trace(dcm) + 1) / 4
    b1_square = (2*dcm[0,0] - np.trace(dcm) + 1) / 4
    b2_square = (2*dcm[1,1] - np.trace(dcm) + 1) / 4
    b3_square = (2*dcm[2,2] - np.trace(dcm) + 1) / 4

    data = np.array([b0_square, b1_square, b2_square, b3_square])
    names = ['b0_square', 'b1_square', 'b2_square', 'b3_square']
    max_index = np.argmax(data)
    result = names[max_index]

    return result

print(largest_val())
```
