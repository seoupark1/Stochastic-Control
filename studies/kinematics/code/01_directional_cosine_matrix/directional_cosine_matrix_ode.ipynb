```python
import numpy as np

def result():

  # tilde 연산자를 사용해서 body frame 기준 각속도를 matrix로 변환, cross product를 행렬곱으로 계산 가능

    omega_1 = np.array([0, -0.3, 0.2])
    omega_2 = np.array([0.3, 0, -0.1])
    omega_3 = np.array([-0.2, 0.1, 0])
    omega_tilde = np.array([omega_1, omega_2, omega_3])

    bn1 = np.array([-0.87097, 0.45161, 0.19355])
    bn2 = np.array([-0.19355, -0.67742, 0.70968])
    bn3 = np.array([0.45161, 0.58065, 0.67742])
    bn = np.array([bn1, bn2, bn3])

    dcm_rate = -omega_tilde @ bn

    return dcm_rate


print(result())
```
