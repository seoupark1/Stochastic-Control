```python
import numpy as np

# (3-2-1) Euler angles to DCM
def EA2C_321(EA_angles):
    EA_angles = np.radians(EA_angles)
    theta1, theta2, theta3 = EA_angles[0], EA_angles[1],EA_angles[2]

    dcm_BN = np.array([[np.cos(theta2)*np.cos(theta1), np.cos(theta2)*np.sin(theta1), -np.sin(theta2)],
                       [np.sin(theta3)*np.sin(theta2)*np.cos(theta1) - np.cos(theta3)*np.sin(theta1), np.sin(theta3)*np.sin(theta2)*np.sin(theta1) + np.cos(theta3)*np.cos(theta1), np.sin(theta3)*np.cos(theta2)],
                       [np.cos(theta3)*np.sin(theta2)*np.cos(theta1) + np.sin(theta3)*np.sin(theta1), np.cos(theta3)*np.sin(theta2)*np.sin(theta1) - np.sin(theta3)*np.cos(theta1), np.cos(theta3)*np.cos(theta2)]])

    return dcm_BN

# initial values
EA_angles = np.array([-10, 10, 5])
omega_N = np.array([0.01, -0.01, 0.01])
inertia_tensor_B = np.array([[10, 1, -1],
                             [1, 5, 1],
                             [-1, 1, 8]])

def angular_momentum(omega_N, inertia_tensor_B):
    omega_B = EA2C_321(EA_angles) @ omega_N
    H = inertia_tensor_B @ omega_B

    return H

print(angular_momentum(omega_N, inertia_tensor_B))
```
