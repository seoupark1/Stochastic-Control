## < Problem 1 >
![problem 1](https://github.com/seoupark1/Stochastic-Control/blob/main/assets/images/kinetics/continuous_systems/continuous_systems_01.png?raw=true)

```python
import numpy as np

# initial values
m1, m2, m3, m4 = 1, 1, 2, 2
M = m1 + m2 + m3 + m4

R1 = np.array([1, -1, 2])
R2 = np.array([-1, -3, 2])
R3 = np.array([2, -1, -1])
R4 = np.array([3, -1, -2])

R1_dot = np.array([2, 1, 1])
R2_dot = np.array([0, -1, 1])
R3_dot = np.array([3, 2, -1])
R4_dot = np.array([0, 0, 1])

def kinetic_energy(m1, m2, m3, m4, R1, R2, R3, R4, R1_dot, R2_dot, R3_dot, R4_dot):
    # CM position
    R_CM = (m1 * R1 + m2 * R2 + m3 * R3 + m4 * R4) / M
    R_CM_dot = (m1 * R1_dot + m2 * R2_dot + m3 * R3_dot + m4 * R4_dot) / M

    # kinetic energy
    energy_trans = 1/2 * M * np.vdot(R_CM_dot, R_CM_dot)
    energy_rot = 1/2 * (m1 * np.vdot(R1_dot - R_CM_dot, R1_dot - R_CM_dot) +
                        m2 * np.vdot(R2_dot - R_CM_dot, R2_dot - R_CM_dot) +
                        m3 * np.vdot(R3_dot - R_CM_dot, R3_dot - R_CM_dot) +
                        m4 * np.vdot(R4_dot - R_CM_dot, R4_dot - R_CM_dot))
    return energy_trans, energy_rot

print(kinetic_energy(m1, m2, m3, m4, R1, R2, R3, R4, R1_dot, R2_dot, R3_dot, R4_dot))
```

```python
import numpy as np

# initial values
m1, m2, m3, m4 = 1, 1, 2, 2
R1_dot = np.array([2, 1, 1])
R2_dot = np.array([0, -1, 1])
R3_dot = np.array([3, 2, -1])
R4_dot = np.array([0, 0, 1])

def linear_momentum():
    p = m1 * R1_dot + m2 * R2_dot + m3 * R3_dot + m4 * R4_dot

    return p

print(linear_momentum())
```

## < Problem 3 >
![problem 3](https://github.com/seoupark1/Stochastic-Control/blob/main/assets/images/kinetics/continuous_systems/continuous_systems_03.png?raw=true)

```python
import numpy as np

# initial values
m1, m2, m3, m4 = 1, 1, 2, 2
M = m1 + m2 + m3 + m4

R1 = np.array([1, -1, 2])
R2 = np.array([-1, -3, 2])
R3 = np.array([2, -1, -1])
R4 = np.array([3, -1, -2])

R1_dot = np.array([2, 1, 1])
R2_dot = np.array([0, -1, 1])
R3_dot = np.array([3, 2, -1])
R4_dot = np.array([0, 0, 1])

def angular_momentum(m1, m2, m3, m4, R1, R2, R3, R4, R1_dot, R2_dot, R3_dot, R4_dot):
    # cm position
    R_CM = (m1 * R1 + m2 * R2 + m3 * R3 + m4 * R4) / (m1 + m2 + m3 + m4)
    R_CM_dot = (m1 * R1_dot + m2 * R2_dot + m3 * R3_dot + m4 * R4_dot) / (m1 + m2 + m3 + m4)

    # angular momentum
    H_orbital = np.cross(R_CM, R_CM_dot) * M
    HC = (np.cross(R1 - R_CM, R1_dot - R_CM_dot) * m1
         + np.cross(R2 - R_CM, R2_dot - R_CM_dot) * m2
         + np.cross(R3 - R_CM, R3_dot - R_CM_dot) * m3
         + np.cross(R4 - R_CM, R4_dot - R_CM_dot) * m4)
    H0 = H_orbital + HC

    return H0, HC

print(angular_momentum(m1, m2, m3, m4, R1, R2, R3, R4, R1_dot, R2_dot, R3_dot, R4_dot))
```
