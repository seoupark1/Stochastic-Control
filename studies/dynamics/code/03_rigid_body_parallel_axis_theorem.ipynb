```python
import numpy as np
from attitude import ea321_to_dcm, skew_symmetric

# initial values
M = 12.5
euler_angles_deg = np.array([-10, 10, 5])
euler_angles_rad = np.radians(euler_angles_deg)
Ic_B = np.array([[10, 1, -1],
                [1, 5, 1],
                [-1, 1, 8]])
dcm_bn = ea321_to_dcm(euler_angles_rad)

# change Rcp_N to Rcp_B
Rcp_N = np.array([-0.5, 0.5, 0.25])
Rcp_B = dcm_bn @ Rcp_N

def parallel_axis(Ic_B, M, Rcp_B):
    Ip_B = Ic_B + M * skew_symmetric(Rcp_B) @ np.transpose(skew_symmetric(Rcp_B))

    return Ip_B

print(parallel_axis(Ic_B, M, Rcp_B))
```
